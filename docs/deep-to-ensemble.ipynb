```python
import numpy as np
import deep_to_ensemble as d2e
import pylab as pl
%pylab inline
```

```python
# Plotting code
## Heatmap
def hm(matrix,xlabel="",ylabel="",title=""):
    pl.xlabel(xlabel)
    pl.ylabel(ylabel)
    pl.title(title)
    pl.pcolor(matrix)
    pl.colorbar()
    pl.show()

## lineplot multiple yarrays
def lp(xarray,yarrays,labels,xlabel="",ylabel="",title=""):
    pl.xlabel(xlabel)
    pl.ylabel(ylabel)
    pl.title(title)
    for yarray,label in zip(yarrays,labels):
        pl.plot(xarray,yarray,'o-',label=label)
    pl.legend()
    pl.show()
```

```python
#############################################################
## experiment 1 -- Given a single deep network, 
## sweep all values of L and K, calculate M:
##
M_S = 32
L_S = 32
S = d2e.net(M=M_S, L=L_S) # single deep network
##
KLMAnalytical = np.zeros((L_S+1,L_S+1))
KLM = np.zeros((L_S+1,L_S+1))
KLNumParam = np.zeros((L_S+1,L_S+1))
KLNumParamNorm = np.zeros((L_S+1,L_S+1))

temp_net = d2e.net(M=M_S, L=L_S)
for K in range(1,L_S+1):
    for L in range(1,L_S+1):
        
        # Given K,L, and the deep network, compute M
        KLM[K,L] = d2e.getM_L(S,K,L) 
        KLMAnalytical[K,L] = d2e.getM_L_analytical(S,K,L)
        
        # Compute total number of parameters in the ensemble
        temp_net.M = KLM[K,L]
        temp_net.L = L
        KLNumParam[K,L] = K*temp_net.total()
        
        # Exception when M cannot be computed
        if KLM[K,L] == -1:
            KLNumParam[K,L] = 0

# Normalize KLNumParam
KLNumParamNorm = KLNumParam/np.max(KLNumParam)
heatmap = True
lineplot = True
if heatmap:       
    hm(KLMAnalytical,xlabel="K",ylabel="L",title="KLMAnalytical")
    hm(KLM,xlabel="K",ylabel="L",title="KLM")
    hm(KLNumParam,xlabel="K",ylabel="L",title="KlNumParam")
    hm(KLNumParamNorm,xlabel="K",ylabel="L",title="KLNumParamNorm")
if lineplot:
    yarrays = [KLM[2,:][1:], KLM[4,:][1:], KLM[8,:][1:], KLM[16,:][1:], KLM[32,:][1:] ]
    labels = ['K=2','K=4','K=8','K=16','K=32']
    lp(range(1,33),yarrays,labels,xlabel="L",ylabel="M")
#############################################################

```

```python
#############################################################
## experiment 2 -- Given a single deep network, 
## sweep all values of M and K, calculate L:
##
M_S = 32
L_S = 32
S = d2e.net(M=M_S, L=L_S) # single deep network
##
KML = np.zeros((M_S+1,M_S+1))
KMNumParam = np.zeros((M_S+1,M_S+1))
KMNumParamNorm = np.zeros((M_S+1,M_S+1))

temp_net = d2e.net(M=M_S, L=L_S)
for K in range(1,M_S+1):
    for M in range(1,M_S+1):
        
        # Given K, M, and the deep network compute L
        KML[K,M] = d2e.getL_M(S,K,M)
        
        # Compute the total number of param of this config
        temp_net.M = M
        temp_net.L = KML[K,M]
        KMNumParam[K,M] = K*temp_net.total()
        
        # exception when L cannot be computed
        if KML[K,M] == -1:
            KMNumParam[K,M] = 0

# Normalize KLNumParam
KMNumParamNorm = KMNumParam/np.max(KMNumParam)
heatmap = True
lineplot = True
if heatmap:       
    hm(KML,xlabel="K",ylabel="L",title="KLM")
    hm(KMNumParam,xlabel="K",ylabel="L",title="KlNumParam")
    hm(KMNumParamNorm,xlabel="K",ylabel="L",title="KLNumParamNorm")
if lineplot:
    yarrays = [KML[2,:][1:], KML[4,:][1:], KML[8,:][1:], KML[16,:][1:], KML[32,:][1:] ]
    labels = ['K=2','K=4','K=8','K=16','K=32']
    lp(range(1,33),yarrays,labels,xlabel="M",ylabel="L")
#############################################################


```

```python
# 
def getM_L_analytical(S,K,L):
    S_ = S.total()
    
    #
    sqrt_terms = math.sqrt((L*((36*S_*1.0)/K*1.0) + L + 1306) + 693889)
    nominator = sqrt_terms - L - 833
    denominator = 18*L*1.0
    M = nominator/denominator
    return math.floor(M)
```

```python
# 
def getL_M_analytical(S,K,M):
    S_ = S.total()
    #
    nominator = S_ - (833 * K * M) - (10 * K)
    denominator = K * M *( (9 * M) + 1)
    L = nominator*1.0/denominator*1.0
    return math.floor(L)
```
